```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Safari/537.36."})
    soup = bts(result.text, "html.parser")
    return soup
```

```python
page = requests.get(url).text
soup = bts(page, "html.parser")
```

```python
soup.findAll("div", {"class": "mw-allpages-body"})
```

# Sayfadaki Linklerin Elde Edilmesi

```python
href_list = []
for li in li_tags:
    a_tag = li.find("a")
    if a_tag is not None:
        href_list.append(a_tag["href"])
new_href_list = []
for href in href_list:
    new_href = "https://tr.wikipedia.org/" + href
    new_href_list.append(new_href)
for new_href in new_href_list:
    print(new_href)
```

# Tek bir sayfa için döngü oluşturma

```python
from bs4 import BeautifulSoup
from selenium import webdriver
başlık = []
içerik = [] 

def scrape_page(url):
    driver = webdriver.Chrome() 
    driver.get(url)

    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    
    span_tag = soup.find("span", {"class": "mw-page-title-main"})
    if span_tag is not None:
        title_text = span_tag.text
        başlık.append(title_text)
    
    div_tag = soup.find("div", {"id": "mw-content-text", "class": "mw-body-content"})
    if div_tag is not None:
        div_text = div_tag.get_text(separator="\n")
        içerik.append(div_text)
        
for href in new_href_list:
    scrape_page(href)
```
